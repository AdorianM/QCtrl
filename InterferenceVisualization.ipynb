```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
from qiskit import execute, Aer

from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def draw_states(states, images):
    megaImage = Image.new('RGB', (mat_size, mat_size), color = 'black')

    d = ImageDraw.Draw(megaImage)

    numbers = np.zeros(mat_size*mat_size)

    for state in range(mat_size*mat_size):
        numbers[state] = (int(states[state], 2))

    images.append(numbers.reshape(mat_size,mat_size))
    
```

```python
# a simple example, where the Hadamard gate is applied only to the first qbit over 8 iterations

image_count = 8
images = []

for i in range(image_count):
    qreg_q = QuantumRegister(3, 'q')
    creg_c = ClassicalRegister(3, 'c')
    circuit = QuantumCircuit(qreg_q, creg_c)

    for x in range(i):
        circuit.h(qreg_q[0])

    circuit.measure(qreg_q[0], creg_c[0])
    circuit.measure(qreg_q[1], creg_c[1])
    circuit.measure(qreg_q[2], creg_c[2])
    
    job = execute(circuit,Aer.get_backend('qasm_simulator'),shots=mat_size*mat_size, memory=True)
    counts = job.result().get_counts(circuit)
    states = job.result().get_memory(circuit)
    
    draw_states(states, images)
    
for i in range(image_count):
    plt.figure()
    plt.imshow(images[i], cmap='hot')
```

```python
# a bit more complex test, where we check different circuits

image_count = 4
images = []

for i in range(image_count):
    qreg_q = QuantumRegister(3, 'q')
    creg_c = ClassicalRegister(3, 'c')
    circuit = QuantumCircuit(qreg_q, creg_c)

    if i == 0:
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.h(qreg_q[2])
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.h(qreg_q[2])
        circuit.measure(qreg_q[0], creg_c[0])
        circuit.measure(qreg_q[1], creg_c[1])
        circuit.measure(qreg_q[2], creg_c[2])
    if i == 1:
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.h(qreg_q[2])
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.measure(qreg_q[0], creg_c[0])
        circuit.measure(qreg_q[1], creg_c[1])
        circuit.measure(qreg_q[2], creg_c[2])
    if i == 2:
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.h(qreg_q[2])
        circuit.h(qreg_q[0])
        circuit.measure(qreg_q[0], creg_c[0])
        circuit.measure(qreg_q[1], creg_c[1])
        circuit.measure(qreg_q[2], creg_c[2])
    if i == 3:
        circuit.h(qreg_q[0])
        circuit.h(qreg_q[1])
        circuit.h(qreg_q[2])
        circuit.measure(qreg_q[0], creg_c[0])
        circuit.measure(qreg_q[1], creg_c[1])
        circuit.measure(qreg_q[2], creg_c[2])

    job = execute(circuit,Aer.get_backend('qasm_simulator'),shots=mat_size*mat_size, memory=True)
    counts = job.result().get_counts(circuit)
    states = job.result().get_memory(circuit)
    
    draw_states(states, images)
    
for i in range(image_count):
    plt.figure()
    plt.imshow(images[i], cmap='hot')
```
